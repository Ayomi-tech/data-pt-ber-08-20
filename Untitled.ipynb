```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
help(plt.plot) 
```

```python
x = [1, 2, 3, 4, 5]
y = [1, 2, 3, 4, 5]

#plt.plot(x, y, 'dg') # u can write the color together with the type of plot u want but its nice to seperate it
# your series must have the same lenght
plt.plot(x, y, 'd', color = 'red')
plt.title('This is a plot')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

# plotting multiple series

```python
# line plot
z = [5, 4, 3, 2, 1]
plt.plot(x, y, color = 'magenta')
plt.plot(x, z, color = 'yellow', linewidth = 5)
plt.legend(['kitty','cat'])
plt.show()
```

# Scatter plot

```python
a = np.random.random(10)
b = np.random.random(10)

plt.scatter(a, b, color = 'green')
plt.xlabel('A')
plt.ylabel('B')
plt.show()
```

# Pie charts

```python
animals = ['Cat', 'Dogs', 'Hamster']
values = [20, 30, 50]

plt.pie(values, labels = animals, colors = ['yellow', 'blue', 'green'])
plt.show()
```

# Bar chart

```python
plt.bar(animals, height = values) # we use height cos we need to assign what our bar should carry(bar is also height)
plt.show()
```

```python
plt.barh(animals, width = values) #we use width cos we are plotting to the side
plt.show()
```

```python
x1 = np.arange(0,10,1)
x2 = np.linspace(0,5,10)
plt.scatter(x1, x2)
plt.show()
```

```python
df = pd.read_csv('vehicles.csv')
df.head(2)
```

```python
df.sort_values(by=['City MPG','Highway MPG' ], inplace=True)
plt.plot(df['City MPG'],df['Highway MPG'])
plt.show()
```

```python
df.plot(x = 'City MPG', y='Highway MPG' )
plt.show()
```

```python
group=df[['City MPG', 'Year']].groupby(['Year']).mean().reset_index()
group
group.plot(x = 'Year', y= 'City MPG')        #we use group to plot because it carring our aggregation.
plt.show()
```

```python
# you can also plot by plt because we have to include the variable of your aggregation code and we have to specify the x and y 
plt.plot(group['Year'], group['City MPG'])
plt.show()
```

# multiple line plot: average City MPG and Highway MPG by year


```python
group2 = df[['City MPG','Highway MPG', 'Year']].groupby(['Year']).mean().reset_index()
plt.plot(group2['Year'], group2[['City MPG','Highway MPG']])
plt.show()
```

```python
group2.plot.bar(x ='Year', y = ['City MPG','Highway MPG'])
plt.show()
```
