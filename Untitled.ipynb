# Linear Regression

```python
x=df['HeadLength']
y=df['BodyLength']
# adding constant 
x = sm.add_constant(x)
```

```python
# creating a model
model = sm.OLS(y,x)
model
```

```python
# fitting the model
results=model.fit()
print(results)
```

```python
results.summary()
```

```python
y_pred = results.predict(x)
```

```python
residuals =y -y_pred
```

```python
residuals
```

```python
np.mean(residuals*residuals)
```

```python
plt.plot(x['HeadLength'], residuals, 'o')
```
