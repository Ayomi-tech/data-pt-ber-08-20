```python
import pandas as pd
import numpy as np
import seaborn as sns
import statsmodels.api as sm
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
pd.set_option("display.max_columns",None)
```

```python
logs_web =pd.read_csv('logs_web.csv')
logs_web.head()
```

```python
logs_web.shape
```

```python
logs_not = pd.read_csv('logs_notifications.csv')
logs_not.head()
```

```python
logs_not.shape
```

```python
# Merging 
results = pd.merge(logs_not,logs_web, how='left', on=['ip_anonymized'])
results.head(60)
```

```python
results.isnull().sum(axis = 0)
```

```python
# np.where(df['column name'].isnull(), 0,1)  filipa's code 
results["browser_y"].fillna(0, inplace = True)
```

```python
results['browser_y'].value_counts()
```

```python
# creating another column, if ['browser_y']isnot 'NAN' give me 1 ir not give me 0
results['proceed'] = np.where(results['browser_y']==0,0,1)
results.head()
```

```python
results['proceed'].value_counts()
```

```python
results['message_x'].value_counts()
```

```python
x = pd.get_dummies(results['message_x'],drop_first=True)
x
```

```python
Y = results['proceed']
X = x
X = sm.add_constant(X)
model = sm.Logit(Y, X)
results = model.fit()
results.summary()
```

```python
Y = results['proceed']
X = results['browser_x'] 
X = sm.add_constant(X)
model = sm.Logit(Y, X)
results = model.fit()
results.summary()
```
